# Comparisons with different lengths

```python
%matplotlib notebook

max_length = 5000

lengths = range(max_length)
byte_lengths = [4,16,64,256]
numbers = ['1'+''.join(random.choices(['0', '1'], k=length)) for length in lengths]
byte_based = [[len(byte_based_code(number, byte_length)) for number in numbers] for byte_length in byte_lengths]
elias = [len(elias_code(number)) for number in numbers]

plt.figure(figsize=(10,7))
plt.plot(elias, label='Elias code')
for length, data in zip(byte_lengths, byte_based):
    plt.plot(data, label=f'Byte = {length}')

plt.plot(lengths, label='Identity (upper bound)')
    
plt.legend(loc=2)
plt.ylabel('Code length')
plt.xlabel('Binary number length')
```
